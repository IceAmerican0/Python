```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url="https://www.rottentomatoes.com/top/bestofrt/?year=2019"
page=urlopen(url)

titles=soup.select(".table a")
ranks=soup.select(".table  span.tMeterScore")
reviews=soup.select(".table  td.right.hidden-xs")

count=0

titleLists=[]
countLists=[]
rankLists=[]
reviewLists=[]

for i in range(len(titles)):
    count += 1
    countLists.append(count)
    titleLists.append(titles[i].string.strip())
    rankLists.append(ranks[i].string.strip())
    reviewLists.append(reviews[i].string.strip())
   
df = pd.DataFrame(data=countLists, columns=['Rank'])
df['Rating'] = pd.Series(rankLists)
df['Titles'] = pd.Series(titleLists)
df['Review'] = pd.Series(reviewLists)
df

df.to_csv("./Data/RottenTomatoesList1.csv",sep=",",encoding='utf-8',index=False)
```
