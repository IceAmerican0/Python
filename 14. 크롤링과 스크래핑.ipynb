```python
from bs4 import BeautifulSoup
html="""
    <html>
        <body>
            <h1>Header</h1>
            <p>Line 1을 서술</p>
        </body>
    </html>
"""

soup=BeautifulSoup(html,'html')

h1=soup.html.body.h1
p1=soup.html.body.p

print(h1)
print(p1)
print('-----------')
```

```python
html="""
    <html>
        <body>
            <h1 id="title">Header</h1>
            <p id="body">Line 1을 서술</p>
        </body>
    </html>
"""

soup=BeautifulSoup(html,"html.parser")

title=soup.find(id='title')
body=soup.find(id='body')

print(title)
print(body)
print("title = ",title.string)
print("body = ",body.string)
```

```python
html="""
    <html>
        <body>
            <ul>
                <li><a href="http://www.naver.com">naver</a></li>
                <li><a href="http://www.daum.net">daum</a></li>
            </ul>
        </body>
    </html>
"""

soup=BeautifulSoup(html,'html.parser')
links=soup.find_all('a')
for link in links:
#    print(link.string)
#    print(link.attrs['href'])
    href=link.attrs['href']
    text=link.string
    print(text,">>>",href)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

url="http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

res=req.urlopen(url)
soup=BeautifulSoup(res,"html.parser")

title=soup.find('title').string
print(title)
wf=soup.find('wf').string
for i in wf.split('<br />'):
    print(i)
    
list_wfs=list(wf)
except_char=['<','b','r',"/",">"]
result=""

for lwf in list_wfs:
    if lwf not in except_char:
        result+=lwf
        
print('-'*100)
print(result)
type(result)
result.split('o')
```

# css 선택자 사용하기

soup.select_one() : css선택자로 요소 하나 추출.
soup.select() : css선택자로 요소 여러개를 리스트로 추출.

```python
html="""
    <html>
        <body>
            <div id="meigen">
                <h1>위키 북스</h1>
                <ul class="items">
                    <li>유니티 게임 이펙트 입문서</li>
                    <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
                    <li>모던 웹사이트 디자인의 정석</li>
                </ul>
            </div>
        </body>
    </html>
"""

soup=BeautifulSoup(html,"html.parser")

h1=soup.select_one("div#meigen > h1").string # id:#, class:., >:자손, " ":후손
print(h1)

li_lists=soup.select("div#meigen > ul.items>li")
#print(li_lists)
for li in li_lists:
    print(li.string)
```

```python
#https://finance.naver.com/marketindex/

url="https://finance.naver.com/marketindex/"
res=req.urlopen(url)

soup=BeautifulSoup(res,"html.parser")
prices=soup.select("div.head_info>span.value")

# for price in prices:
#    print(price.string)
print("미국 : ",prices[0].string)
print("일본 : ",prices[1].string)
print("유럽연합 : ",prices[2].string)
print("중국 : ",prices[3].string)
```

```python
url="https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res=req.urlopen(url)

soup=BeautifulSoup(res,"html.parser")

poems=soup.select("div.mw-parser-output>ul>li a")

for poem in poems:
    href =  'https://ko.wikisource.org/' + poem.attrs['href']
    text = poem.string
    if text=="증보판":
        continue
    print(text, ">>>", href)
    
```

```python
url="https://movie.daum.net/ranking/boxoffice/yearly?date=2020"
res=req.urlopen(url)

soup=BeautifulSoup(res,"html.parser")

movies=soup.select(" strong > a")

i=1
for movie in movies:
    print("%2d : " %i,movie.string)
    i+=1
```

```python
url="https://news.daum.net/digital#1"
res=req.urlopen(url)

soup=BeautifulSoup(res,"html.parser")

newses=soup.select("#mAside > div.aside_g.aside_ranking a.link_txt")

i=1
for news in newses:
    href=news.attrs['href']
    print("%d위 : "%i,href," : ",news.string)
    i+=1
```
